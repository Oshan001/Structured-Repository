## 1) Working With dictionary

# 1.1) Creating Dictionary

```python
student = {
    "name" : "oshan",
    "roll_no" : 230328,
    "age" : 22,
    "facality" : "BCE"

}
```

# 1.2) Access value of Dictionary

```python
print (student["name"])
```

# 1.2) Use Get

```python
student.get("name")
```

# 1.3) Add item

```python
student["phone_no"] = 97416635741
student
```

# 1.4) Update value

```python
student["roll_no"]= 28
student
```

# 1.5) Delete value using del

```python
del student["phone_no"]
student
```

# 1.6) Remove using pop

```python
student.pop("facality")
student
```

# 1.7) Automatic remove item using popitem(remove from last)

```python
student.popitem()
student
```

# 1.8) Copy Dict

```python
copyStudent = student.copy()
copyStudent
```

# 1.9)Get All Keys

```python
student.keys()
```

# 1.10)Get All Values

```python
student.values()
```

# 1.11) Clear Dictionary

```python
copyStudent.clear()
copyStudent
```

# 1.12) Get All Items

```python
student.items()
```

# 1.13) Check Key Exists

```python
"name" in student
```

# 1.13) Get Length of Dictionary

```python
len(student)
```

# 1.14) Loop Through Keys

```python
for k in student :
    print(k)
```

# 1.15) Loop Through Values

```python
for i in student.values():
    print(i)
```

# 1.16)Loop Through Items

```python
for i,j in student.items():
    print(i,j)
```
