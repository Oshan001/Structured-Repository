# One-Page Summary: Logistic Regression, SVC, and SVR Comparison

## 1. Logistic Regression (Classification Model)
Logistic Regression is a linear classification algorithm that predicts the probability of a class using the sigmoid function. It is mainly used for binary and multi-class classification problems.

- **Type:** Linear classifier  
- **Output:** Discrete class labels (0/1 or multi-class)  
- **Strengths:**
  - Simple and fast
  - Easy to interpret
  - Works well for linearly separable data  
- **Limitations:**
  - Poor performance on non-linear data
  - Requires feature scaling for best performance  

---

## 2. Support Vector Classifier (SVC)
SVC is a powerful classification algorithm that finds the optimal hyperplane that maximizes the margin between classes. It supports different kernels to handle both linear and non-linear problems.

- **Type:** Classification model  
- **Output:** Class labels  
- **Kernels Used:**
  - Linear kernel
  - RBF kernel  
- **Strengths:**
  - High accuracy on complex datasets
  - Works well in high-dimensional spaces
  - Kernel trick enables non-linear classification  
- **Limitations:**
  - Computationally expensive on large datasets
  - Requires tuning of parameters (C, gamma)

---

## 3. Support Vector Regression (SVR)
SVR is the regression version of SVM. Instead of predicting classes, it predicts continuous values such as housing prices.

- **Type:** Regression model  
- **Output:** Continuous numerical values  
- **Kernels Used:**
  - Linear SVR
  - RBF SVR  
- **Strengths:**
  - Handles non-linear regression well
  - Robust to outliers using epsilon-insensitive loss
  - Flexible with kernel selection  
- **Limitations:**
  - Sensitive to hyperparameters (C, epsilon, gamma)
  - Slower training on large datasets  

---

## Overall Comparison

| Model | Task Type | Output | Strength | Weakness |
|------|----------|--------|----------|----------|
| Logistic Regression | Classification | Class labels | Simple, interpretable | Limited for non-linear patterns |
| SVC | Classification | Class labels | High accuracy, flexible kernels | Computational cost |
| SVR | Regression | Continuous values | Flexible, robust | Requires tuning |
