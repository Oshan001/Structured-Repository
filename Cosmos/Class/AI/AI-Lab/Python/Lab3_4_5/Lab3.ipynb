```python
s = 0
bill = str()
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    price = int(input("Enter price = "))
    quantity = int(input("Enter quantity = "))
    total=price*quantity
    bill = bill + f"{name} {price} {quantity} {total}\n"
    s = s+total

print("\nall total = ",s)
print(bill)
```
