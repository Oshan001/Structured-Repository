```python
#  Water Jug Problem
print('Water Jug Problem')
X= int(input('enter X: '))
Y= int(input('enter Y: '))
while 1:
    rno = int(input('enter Rules NO: '))
    if rno == 1:
      if X < 4 :X=4
    if rno == 2:
      if Y < 3 :Y=3
    if rno == 5:
        if X > 0: X = 0
    if rno == 6:
        if Y > 0: Y = 0
    if X + Y >= 4 and Y > 0: X, Y = 4, Y - (4 - X)
    if X + Y >= 3 and X > 0: X, Y = X - (3 - Y), 3
    if rno == 9:
        if X + Y <= 4 and Y > 0: X, Y = X + Y, 0
    if rno == 10:
        if X + Y <= 3 and X > 0: X, Y = 0, X + Y
        print("X =", X)
    print("Y =", Y)
    if X == 2:
        print(" The result is a Goal state")
    break


```

```python
# Breadth First Search
graph = {
 'A' : ['B','C'],
 'B' : ['D', 'E'],
 'C' : ['F'],
 'D' : [],
 'E' : ['F'],
 'F' : []
}
visited = []
queue = []
def bfs(visited, graph, node):
 visited.append(node)
 queue.append(node)
 while queue:
   S = queue.pop(0)
   print (S, end = " ")
 for neighbour in graph[S]:
    if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)
#output
bfs(visited, graph, 'C')
```
