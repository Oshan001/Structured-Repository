```python
# break
i = 0
while i<10:
    if i ==6:
        break
    print(i,end = "")
    i = i+1
```

```python
# continue
i = 0
while i<10:
    if i ==6:
        i = i+1
        continue
    print(i,end = "")
    i = i+1
```
