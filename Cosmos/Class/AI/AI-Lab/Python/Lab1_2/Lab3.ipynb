```python
phy = int(input("enter mark of physics ="))
chem = int(input("enter mark of chemistry ="))
math = int(input("enter mark of math ="))
eng = int(input("enter mark of english ="))
nepali = int(input("enter mark of nepali ="))

total = phy + chem + math + eng + nepali
per = total/5
if per>= 80:
    grade = "A"
elif per>= 70:
    grade = "B"
elif per>= 60:
    grade = "C"
elif per>= 45:
    grade = "F"
print("total is ",total)
print("per is ",per)
print("grade is ",grade)
```

```python
a = int(input("Enter a = "))
b = int(input("Enter b = "))
o = input("Enter o = ")
if o == '+':
    print(a+b)
elif o == '-':
    print(a-b)
elif o == '*':
    print(a*b)
elif o == '/':
    if b != 0:
        print(a/b)
    else:
        print("The vale of b can not be zero.")
else:
    print("Enter Valid Operator")
```
