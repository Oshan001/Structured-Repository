# Image rotation

```
import cv2
import imutils
image = cv2.imread('/content/download.jpg')
from google.colab.patches import cv2_imshow
cv2_imshow(image)
print('Original Image')
rotated = imutils.rotate(image, 90)
cv2_imshow(rotated)
print('90 degree rotated image')
```

# Conversion of RGB model to CMY Model





```
import cv2
image = cv2.imread('/content/download.jpg')
from google.colab.patches import cv2_imshow
cv2_imshow(image)
print('Original Image')
R, G, B = cv2.split(image)
C = 255 - R
M = 255 - G
Y = 255 - B
cv2_imshow(R)
print('Red Component')
cv2_imshow(G)
print('Green Component')
cv2_imshow(B)
print('Blue Component')
cv2_imshow(C)
print('Cyan Component')
cv2_imshow(M)
print('Magenta Component')
cv2_imshow(Y)
print('Yellow Component')
```
